```python
import numpy as np
from sklearn import svm
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from flask import Flask,render_template,request
```

```python
data=pd.read_csv('bsdara.csv')
data
```

```python
datcop=data
```

```python
datcop=datcop.drop(['id'],axis=1)
```

```python
training_data=datcop[['radius_mean','radius_worst','radius_se', 'texture_mean',
       'texture_se', 'texture_worst', 'perimeter_mean','perimeter_se', 'perimeter_worst','area_mean', 'area_se','area_worst']]
training_data.head(10)
```

```python
output_data=datcop['diagnosis']
```

```python
X, x_test, Y, y_test = train_test_split(training_data, output_data, test_size=0.2, random_state = 42)
```

```python
print(X)
```

```python
model1=svm.SVC(kernel='linear',C=1,gamma=1)
model2=DecisionTreeClassifier()
model3=KNeighborsClassifier(n_neighbors=5)
```

```python
evc=VotingClassifier(estimators=[('model1',model1),('model2',model2),('model3',model3)])
```

```python
evc.fit(X,Y)
```

```python
prediction = evc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
score = accuracy_score(y_test, prediction)
score
```

```python
app=Flask(__name__)
```

```python
@app.route('/')
def hello():
    return render_template('index.html')
@app.route('/send',methods=['GET','POST'])
def send():
    if request.method =='POST':
        Rmean=request.form['Rmean']
        Rse=request.form['Rse']
        Rworst=request.form['Rworst']
        Tmean=request.form['Tmean']
        Tse=request.form['Tse']
        Tworst=request.form['Tworst']
        Pmean=request.form['Pmean']
        Pse=request.form['Pse']
        Pworst=request.form['Pworst']
        Amean=request.form['Amean']
        Ase=request.form['Ase']
        Aworst=request.form['Aworst']
        lis1=request.form
        lis=[Rmean,Rse,Rworst,Tmean,Tse,Tworst,Pmean,Pse,Pworst,Amean,Ase,Aworst]
        val=evc.predict(np.array(lis).reshape(1, -1))
        if(val[0]=='B'):
            return render_template('result.html',val='Tumor is benign')
        else:
            return render_template('result.html',val='Tumor is Malignant')
    return render_template('index.html')

def shutdown_server():
    func = request.environ.get('werkzeug.server.shutdown')
    if func is None:
        raise RuntimeError('Not running with the Werkzeug Server')
    func()

@app.route('/shutdown', methods=['POST'])
def shutdown():
    shutdown_server()
    return 'Server shutting down...'
        
app.run()
```
